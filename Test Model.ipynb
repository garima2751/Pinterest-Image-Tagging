```python
import os
import numpy as np
import keras
from keras.models import load_model
from keras.applications.mobilenet import preprocess_input
from keras.preprocessing import image
import pandas as pd
import cv2
```

```python
labels_to_names = {0 : 'animals', 1 : 'art', 2 : 'automobiles', 3 : 'fashion', 4 : 'food', 5 : 'houses',
                   6 : 'technology', 7 : 'travel'}
```

```python
from keras.applications.mobilenet import preprocess_input
```

```python
model = load_model('D:/Ryerson University/Courses/DS - 8013 Deep Learning/Project/bin/snapshots/MobileNet/model_mobilenet_lr0.0001_bs32_12.hdf5')

df = pd.DataFrame(columns = ['animals', 'art', 'automobiles', 'fashion', 'food', 'houses', 'technology', 'travel'], 
                  index = ['animals', 'art', 'automobiles', 'fashion', 'food', 'houses', 'technology', 'travel'])
df[:] = 0
total_count = 0
correct_count = 0

for i in os.listdir('D:/Ryerson University/Courses/DS - 8013 Deep Learning/Project/bin/dataset/test_set'):
    true_class = i
    for j in os.listdir('D:/Ryerson University/Courses/DS - 8013 Deep Learning/Project/bin/dataset/test_set/' + i):
        test_image = image.load_img('D:/Ryerson University/Courses/DS - 8013 Deep Learning/Project/bin/dataset/test_set/' + i + '/' + j, 
                                    target_size = (160,160))
        test_image = image.img_to_array(test_image)
        test_image = preprocess_input(test_image)
        test_image = np.expand_dims(test_image, axis = 0)
        result = model.predict(test_image)
        
        prediction = np.argmax(result)
        
        prediction = labels_to_names[prediction]
        print(true_class + ' : ', prediction)
        df.loc[true_class, prediction] += 1
        total_count += 1
        if true_class == prediction:
            correct_count += 1

print(df)

avg_accuracy = round((correct_count * 100) / (total_count), 3)
print("\nAverage Model Accuracy on Test Set = " + str(avg_accuracy) + "%")
```
